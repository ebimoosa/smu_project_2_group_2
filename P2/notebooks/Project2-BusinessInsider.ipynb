```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import requests
import os
from sqlalchemy import create_engine
import numpy as np
import matplotlib.pyplot as plt
from config import key
import psycopg2
```

```python
url = 'https://www.businessinsider.com/best-video-games-metacritic-2017-11#50-world-of-goo-2'
response = requests.get(url)
html = response.text
html[0:1000]
```

```python
game_soup = bs(html, 'lxml')
```

```python
game_list = game_soup.find_all("div", {"class": "slide"})
game = game_list[0]
print(game)
```

```python
games = game_soup.find_all("div", {"class": "slide-layout clearfix"})
game = games[1]
game
```

```python
name = game.find("h2", {"class": "slide-title-text"}).text
name = name[4:-1]
name
```

```python
name = game.find("h2", {"class": "slide-title-text"}).text[3:].strip('""')
name
```

```python
critic = game.find("p").text.strip("Critic score:")
critic = critic[-6:3]
critic
```

```python
critic = game.find("p").text.strip("Critic score:")[-6:-4]
critic
```

```python
user = game.find_all("p")[1].text.strip("User score:")
user = user[-6:-3]
user
```

```python
data = []

for game in games:
    name = game.find("h2", {"class": "slide-title-text"}).text[4:-1].strip('""')
    
    critic = game.find("p").text.strip("Critic score:")[-6:-4]
    user = game.find_all("p")[1].text.strip("User score:")[-6:-3]
   
    
    info = {"Name":name, "Critic Score":critic, "User Score": user}
    data.append(info)
```

```python

for x in range(1, 51):
    print(data[x]) 
```

```python
df = pd.DataFrame(data)
df.head()
```
