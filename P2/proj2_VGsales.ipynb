```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import requests
import os
```

```python
url = 'https://vgsales.fandom.com/wiki/List_of_best-selling_video_games'
response = requests.get(url)
html = response.text
html[0:1000]
```

```python
game_soup = bs(html, 'lxml')
```

```python
game_list = game_soup.find_all("div", {"class": "mw-parser-output"})
game = game_list[0]
print(game)
```

```python
tables = game.find_all("tr")
table = tables[1]
table
```

```python
name = table.find('td').text.strip("\n")
name
```

```python
genre = table.find_all("td")[2].text.strip("\n")
genre
```

```python
year = table.find_all("td")[1].text.strip("\n")
year
```

```python
platform = table.find_all("td")[3].text.strip("\n")
platform
```

```python
unit = table.find_all("td")[4].text.strip("\n")
unit
```

```python
data = []

for table in tables:
    
    try:
    
        name = table.find('td').text.strip("\n")
        genre = table.find_all("td")[2].text.strip("\n")
        year = table.find_all("td")[1].text.strip("\n")
        
        unit = table.find_all("td")[4].text.strip("\n")
    except Exception:
        pass
     
    info = {"Game":name, "Genre":genre, "Year Released":year, "Units Sold (millions)":unit}
    data.append(info)
```

```python
for x in range(0,20):
    print(data[x]) 
```

```python
df = pd.DataFrame(data)
df.head(50)
```

```python
df["Game"].nunique()
```

```python
df2 = df.drop_duplicates(subset=["Game"]).reset_index()
df2 = df2.drop('index', 1)
df_clean = df2.head(51)
df_clean
```
